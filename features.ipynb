```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cols = ["shop_kind", "city", "shop_brand"]
shops = pd.read_csv("./sales/_shops.csv", usecols=["shop_id", "shop_kind", "city", "shop_brand"])

for col in cols:
    value_map = dict((v,i) for i,v in enumerate(shops[col].unique()))
    shops = shops.replace({col:value_map})

shops.head()
```

```python
cols = ["category_1", "category_2"]
item_categories = pd.read_csv("./sales/_item_categories.csv", usecols=["item_category_id", "category_1", "category_2"])

for col in cols:
    value_map = dict((v,i) for i,v in enumerate(item_categories[col].unique()))
    item_categories = item_categories.replace({col:value_map})

item_categories.head()
```

```python
items = pd.read_csv("./sales/items.csv")
items
```

```python
feature_count = 25
tfidf = TfidfVectorizer(max_features=feature_count)
items_df_item_name_text_features = tfidf.fit_transform(items['item_name']).toarray().tolist()

items["item_names_tf_idf"] = items_df_item_name_text_features

items.drop('item_name', axis = 1, inplace = True)

items
```
